```python
# Python program to illustrate Support Vector Machine (SVM)
# Import necessary modules 
import numpy as np
import seaborn as sn
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn import metrics 
from sklearn.metrics import classification_report, confusion_matrix
```

```python
# Load data 
data = pd. read_csv('C://Users//Dell//Desktop//Medhack//final_dataset1.csv')
print(data.shape)
print ("sampe data")
print(data.head())
df = pd.DataFrame(data,columns=['SBP','DBP','Pulse','Temperature','Hemoglobin','Age','BMI','Sex','Smoking','alcohol_consumption','Level_of_Stress'])
corrMatrix = df.corr()
sn.heatmap(corrMatrix, annot=True)
plt.show()
```

```python
# Create feature and target arrays 
#Health condition
x1=data.iloc[:,:-7].values
x1
df1 = pd.DataFrame(data,columns=['SBP','DBP','Pulse','Temperature','Hemoglobin','Age','BMI','Sex','Level_of_Stress','Health condition'])
corrMatrix = df1.corr()
sn.heatmap(corrMatrix, annot=True)
plt.show()
```

```python
#Chronic kidney disorder
x2=x=data[['Age','salt_content_in_the_diet','Sex','Smoking','alcohol_consumption']].values
x2
df2 = pd.DataFrame(data,columns=['Age','salt_content_in_the_diet','Sex','Smoking','alcohol_consumption','Chronic_kidney_disease'])
corrMatrix = df2.corr()
sn.heatmap(corrMatrix, annot=True)
plt.show()
```

```python
#Adrenal and tyroid
x3=x=data[['Age','Sex','Level_of_Stress','BMI']].values
x3
df3 = pd.DataFrame(data,columns=['Age','Sex','Level_of_Stress','BMI','Adrenal_and_thyroid_disorders'])
corrMatrix = df3.corr()
sn.heatmap(corrMatrix, annot=True)
plt.show()
```

```python
y1=data.iloc[:,14:15].values
y1
y2=data.iloc[:,12:13].values
y2
y3=data.iloc[:,13:14].values
y3
```

```python
X1_train, X1_test, y1_train, y1_test = train_test_split(x1, y1, test_size=0.33,random_state=10)
print(X1_train)
y1_train = np.reshape(y1_train, (1,np.product(y1_train.shape)))
y1_train = y1_train.ravel()
print(y1_train)
```

```python
X2_train, X2_test, y2_train, y2_test = train_test_split(x2, y2, test_size=0.33,random_state=10)
print(X2_train)
y2_train = np.reshape(y2_train, (1,np.product(y2_train.shape)))
y2_train = y2_train.ravel()
print(y2_train)
```

```python
X3_train, X3_test, y3_train, y3_test = train_test_split(x3, y3, test_size=0.33,random_state=10)
print(X3_train)
y3_train = np.reshape(y3_train, (1,np.product(y3_train.shape)))
y3_train = y3_train.ravel()
print(y3_train)
```

```python
#Import SVM model
svclassifier1 = SVC(kernel='linear')
svclassifier1.fit(X1_train, y1_train)
svclassifier2 = SVC(kernel='linear')
svclassifier2.fit(X2_train, y2_train)
svclassifier3 = SVC(kernel='linear')
svclassifier3.fit(X3_train, y3_train)
```

```python
#Train the model using the training sets
y1_pred = svclassifier1.predict(X1_test)
# Model Accuracy
print("Accuracy:",metrics.accuracy_score(y1_test, y1_pred))
y2_pred = svclassifier2.predict(X2_test)
# Model Accuracy
print("Accuracy:",metrics.accuracy_score(y2_test, y2_pred))
y3_pred = svclassifier3.predict(X3_test)
# Model Accuracy
print("Accuracy:",metrics.accuracy_score(y3_test, y3_pred))
```

```python
#Model Evaluation: Compute classification_report and  confusion_matrix
print(confusion_matrix(y1_test,y1_pred))
print(classification_report(y1_test,y1_pred))
print(confusion_matrix(y2_test,y2_pred))
print(classification_report(y2_test,y2_pred))
print(confusion_matrix(y3_test,y3_pred))
print(classification_report(y3_test,y3_pred))
```

```python
#Plot confusion_matrix1
labels = ['0', '1']
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(confusion_matrix(y1_test,y1_pred))
plt.title('Confusion matrix of the Health condiion')
fig.colorbar(cax)
ax.set_xticklabels([''] + labels)
ax.set_yticklabels([''] + labels)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()

#Plot confusion_matrix2
labels = ['0', '1']
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(confusion_matrix(y2_test,y2_pred))
plt.title('Confusion matrix of the Chronic kidney disease')
fig.colorbar(cax)
ax.set_xticklabels([''] + labels)
ax.set_yticklabels([''] + labels)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()

#Plot confusion_matrix3
labels = ['0', '1']
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(confusion_matrix(y3_test,y3_pred))
plt.title('Confusion matrix of the adrenal')
fig.colorbar(cax)
ax.set_xticklabels([''] + labels)
ax.set_yticklabels([''] + labels)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```
